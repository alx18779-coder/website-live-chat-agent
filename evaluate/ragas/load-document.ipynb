```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import MarkdownHeaderTextSplitter
```

```python
# 加载文件
MD_PATH = "./docs/introduction.md"
loader = TextLoader(MD_PATH,encoding="utf-8")
docs = loader.load()

# 使用MarkdownHeaderTextSplitter进行分割
headers_to_split_on = [
    ("##", "Header 2"),
    ("###", "Header 3"),
    ("####", "Header 4"),
]

spliter = MarkdownHeaderTextSplitter(
    headers_to_split_on=headers_to_split_on,
    strip_headers=False,
)

split_docs = spliter.split_text(docs[0].page_content)
print(split_docs)

for chunk in split_docs:
    print(chunk)
    print(len(chunk.page_content))
    print("-"*100)
```

```python
all_chunks = []
for chunk in split_docs:
    all_chunks.append({
        "text":chunk.page_content[0:500],
        "metadata":{
            "source_file": MD_PATH,
            "title": chunk.metadata.get("title", ""),
            **chunk.metadata,
        }
    })
for i in all_chunks:
    print(i)
    print("-"*100)

```
